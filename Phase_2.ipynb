### 2.1	Analysis of Literacy Rate vs Total Crimes.

Let's analyse Total Crimes over the years.

```python
crimes = df.groupby('Year').sum().reset_index()[['Year', 'Total']]
crimes
```

Here, we can see that the number of crimes has increased 3 times from 2001 to 2011.

```python
#Let's visualize it
plt.figure(figsize=(6,6), facecolor='White')
plt.title("Total Crimes over the years")
sns.lineplot(x = "Year", y = "Total", data=df)
plt.grid()
plt.show()
```

From the above plot we can see the increase in crimes with increase in years.

```python
#Let's see the increase in literacy rate over the years
literacy = df.groupby('Year').sum().reset_index()[['Year', 'Average Literacy Rate %']]
literacy
```

Here, we can see that there is an increase in literacy rates over the years.

```python
#Let's visualize it
plt.figure(figsize=(6,6), facecolor='White')
plt.title("Literacy rate over the years")
sns.lineplot(x = "Year", y = "Average Literacy Rate %", data=df)
plt.grid()
plt.show()
```

From the above plot we can see that literacy rate increases with the increase in years.

From the above plot we can see that males are more literate than females.

In most parts of India, the females are not given the right to education, mostly after they have gained maturity. They are married off at an early age and carry child soon after that. They face many difficulties for being females and most of the places are highly unsafe for them for which their parents don't allow them to pursue higher education in different places.
That is the reason why females are less literate than males.

#### Let's visualize the relationship between states, crimes and literacy rate.

Before visualization let's categorize the average literacy rates to A, B and C respectively otherwise it will create errors while visualizing.

```python
lit_cat = df.copy()
```

```python
def cre(x):
    if x>=50 and x<65:
        return 'C'
    elif x>=65 and x<80:
        return 'B'
    else:
        return 'A'
```

```python
lit_cat['Average Literacy Rate %'] = lit_cat['Average Literacy Rate %'].apply(cre)
lit_cat
```

We have successfully categorized the values, let's proceed ahead.

```python
plt.figure(figsize=(10,6), facecolor='Cyan')
sns.barplot(x = lit_cat['STATE/UT'], y = lit_cat['Total'], data=df, hue=lit_cat['Average Literacy Rate %'], palette='bright')
plt.xticks(rotation=90)
plt.xlabel("STATES/UT")
plt.ylabel("Total crimes")
plt.title("Crimes vs States vs Literacy Rates")
plt.show()
```

Here, we can see that Madhya Pradesh is having moderate literacy rate but highest crimes of all others, Uttar Pradesh is having lowest literacy rate but high number of crimes after MP. We can see that Maharashtra and Delhi, being one of the most literate have surprisingly high crimes.

Whereas Lakshadweep, Puducherry, Goa, Daman & Diu have low crimes but high literacy rates.

### 2.3	Analysis of year-on-year total crime rate.

```python
lit_rate
```

Let's analyse crimes year wise from 2001-2011 using for loop.

```python
year = [x for x in range(2001,2012)]

for i in year:
    plt.figure(figsize=(12,8))
    yr = lit_rate[lit_rate['Year']==i].sort_values(by='Total',ascending=False)
    sns.lineplot(x=yr['STATE/UT'], y=yr['Total'])
    plt.xticks(rotation=90)
    plt.title("Total crimes in different states in the year {}".format(i))
    plt.grid()
    plt.show()
```

In any case, we can see that Lakshadweep has the least crimes over the years 2001-2011 and Madhya Pradesh, Maharashtra and Uttar Pradesh have the highest crimes.
